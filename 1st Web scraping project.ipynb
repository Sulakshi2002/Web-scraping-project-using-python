# Scraping Data from a Real website +Pandas

```python
from bs4 import BeautifulSoup  
import requests 
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
headers = {"User-Agent": "MyLearningBot/1.0 (https://example.com/contact)"}

page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html')

print(soup.prettify())
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable')[1] # get data from second table
```

```python
# in here we use first table so use below command
```

```python
table = soup.find('table', class_ = 'wikitable sortable') # get data from 1st table
```

```python
world_titles=table.find_all('th')
print(world_titles)
```

```python
titles = [title.text.strip() for title in world_titles]
```

```python
print(titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
